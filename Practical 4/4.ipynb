```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# 1. Import the RealEstate_Prices.csv dataset. Clean column names by removing spaces,
# special characters, or renaming them for clarity.
data = pd.read_csv('Real estate.csv')
data
```

```python
data.columns = [
    'ID',
    'TransactionDate',
    'HouseAge',
    'DistanceToMRT',
    'ConvenienceStores',
    'Latitude',
    'Longitude',
    'HousePricePerUnitArea'
]
data
```

```python
# 2. Handle missing values in the dataset, deciding on an appropriate strategy (e.g.,
# imputation or removal).
data.isnull().sum()
```

```python
# 3. Perform data merging if additional datasets with relevant information are available
# (e.g., neighborhood demographics or nearby amenities).
```

```python
# 4. Filter and subset the data based on specific criteria, such as a particular time period,
# property type, or location.
filtered_df = data[(data['TransactionDate'] >= 2013) & (data['DistanceToMRT'] <= 500)]
filtered_df
```

```python
# 5. Handle categorical variables by encoding them appropriately (e.g., one-hot encoding
# or label encoding) for further analysis.
# if 'column name' in data.columns:
#     df = pd.get_dummies(data, columns=['column name'], drop_first=True)
```

```python
# 6. Aggregate the data to calculate summary statistics or derived metrics such as average
# sale prices by neighborhood or property type.
avg_price_by_stores = data.groupby('ConvenienceStores')['HousePricePerUnitArea'].mean()
print("Average house price per unit area by convenience stores:\n", avg_price_by_stores)
```

```python
# 7. Identify and handle outliers or extreme values in the data that may affect the analysis
# or modeling process.
plt.figure(figsize=(15, 5))
sns.boxplot(data=data[['HouseAge', 'DistanceToMRT', 'HousePricePerUnitArea']])
plt.title("Boxplots to Identify Outliers")
plt.show()

# Handling outliers by capping values beyond the 95th percentile
for column in ['HouseAge', 'DistanceToMRT', 'HousePricePerUnitArea']:
    upper_limit = data[column].quantile(0.95)
    lower_limit = data[column].quantile(0.05)
    data[column] = np.clip(data[column], lower_limit, upper_limit)

print("Data wrangling complete.")
```

```python
outlier_columns = ['HouseAge', 'DistanceToMRT', 'HousePricePerUnitArea']

# Capping outliers at the 5th and 95th percentiles
for column in outlier_columns:
    lower_limit = data[column].quantile(0.05)
    upper_limit = data[column].quantile(0.95)
    data[column] = np.clip(data[column], lower_limit, upper_limit)

# Verify changes by checking summary statistics
data[outlier_columns].describe()
```
